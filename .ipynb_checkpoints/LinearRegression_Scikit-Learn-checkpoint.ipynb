### Fonksiyonlarla oluşturmuş olduğum modelimin, sklearn kütüphanesi kullanılarak nasıl daha verimli şekilde oluşturulabileceğini gösterdiğim proje

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("weight-height.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# x ve y'leri tanımladım
x = df[["Boy"]]
y = df["Kilo"]
```

```python
#verilerin dağılımını incelemek adına görselleştirdim
plt.scatter(x, y, c ="blue")
plt.show()
```

```python
# standardizasyonu Sklearn kütüphanesinin StandardScaler fonksiyonu ile sağlayacağım

ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
 x
```

```python
# doğru bir şekilde ilerlemek adına, train ve test olarak verileri bölmem gerekiyor.
x_train, x_test , y_train, y_test = train_test_split(x,y,train_size = 0.8,random_state = 18)
```

```python
lr = LinearRegression()
model = lr.fit(x_train,y_train)
score = model.score(x_test,y_test)
score
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color="orange", label="Eğitim Verileri")  # Eğitim verilerini turuncu noktalarla çizdirdim
plt.scatter(x_test, y_test, color="blue", label="Test Verileri")  # Test verilerini mavi noktalarla çizdirdim
plt.plot(x_test, y_pred, color="black", label="Tahminler")  # Tahminleri siyah çizgiyle çizdirdim

plt.xlabel("Boy") 
plt.ylabel("Kilo")
plt.title("Model Tahminlerinin Görselleştirilmesi")
plt.legend()
plt.show()
```

```python
# bulmuş olduğum bu modelimin prediction'larını normalde olan değerlerle de kıyaslamak istersem eğer şu kod bloğuyla çözerim

sonuc_karsilastirma_df = pd.DataFrame({"Gerçek Değerler":y_test,"Tahminler":y_pred})

sonuc_karsilastirma_df
```
